### 2. Print all the teams which have started playing between 1930-1980.

```python
#Split the 'Debut' column in 'laliga' DatFrame on '-' into two columns and assign it to a temp DataFrame temp_df
temp_df = laliga['Debut'].str.split('-', n=1, expand= True)

#label the columns in temp_df
temp_df.columns=['A','B']
```

```python
#Convert dtypes of all the columns of temp_df to numeric (int or float)... 
#using apply() function and pd.to_numeric attribute
temp_df = temp_df.apply(pd.to_numeric)

#Check the dtypes of temp_df
temp_df.dtypes
```

```python
#Check the column values using head(10).
temp_df.head(10)
```

```python
#counter to count number of rows retrieved
count=1
```

```python
#By default Strings are left-justified and numbers are right justified.
#Using format() function to left-align and right align the column headings in output
#s- string, <4s,<10s- left-align 4 and 10 spaces, >25s- right-align 25 spaces

print('{:<4s}{:<10s}{:<25s}{:>10s}'.format("#", "Index", "Team", "Debut"))

#Iterate over each row in temp_df
#In if condition, using where function retrieve all rows satisfying the given condition and not having null values
#For these rows print the 'Team' and 'Debut' columns from original 'laliga' DataFrame along with row number and count
#In print statement use the same format used for heading for proper alignment

for row in temp_df.index:
    if (temp_df['A'].where((temp_df['A'] >= 1930) & (temp_df['A'] <= 1980)).notnull()[row]):
        print('{:<4}{:<10}{:<25}{:>10}'.format(count, row, laliga.loc[row,'Team'], laliga.loc[row,'Debut']))
        count+=1
```

### 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.Using the same function, find the team which has maximum and minimum goal difference.
#### Goal_diff_count = GoalsFor - GoalsAgainst

```python
#define a function Goal_diff_count() and pass the temp_df dataframe as a parameter
#Add column 'GoalDifferenceCount' to  find difference of 'GoalsFor' and 'GoalsAgainst'
#drop columns 'GoalsFor' and 'GoalsAgainst'and return the DataFrame

def Goal_diff_count(laliga):
    #Create a 'temp_df' DataFrame from 'laliga', with columns to find GoalDifferenceCOunt
    #Change the datatyes of columns to int using astype() function and check the datatypes
    temp_df = laliga[['Team','GoalsFor','GoalsAgainst']]
    temp_df = temp_df.astype({'GoalsFor': int, 'GoalsAgainst': int})
    temp_df['GoalDifferenceCount'] = temp_df['GoalsFor'] - temp_df['GoalsAgainst']
    temp_df.drop(['GoalsFor', 'GoalsAgainst'],axis=1, inplace=True)
    return temp_df
```

```python
#Call the function Goal_diff_count(), with 'laliga' DatFrame as parameter
#The function returns Teams and their GoalDifferenceCount

Goal_diff_count(laliga)
```

```python
#find position of Maximum Goal Difference count using idmax() and return the team and Max GoalDifferenceCount
Goal_diff_count(laliga).iloc[[(Goal_diff_count(laliga)['GoalDifferenceCount']).idxmax()]]
```

```python
#find position of Minimum Goal Difference count using idmin() and return the team and Min GoalDifferenceCount
Goal_diff_count(laliga).iloc[[(Goal_diff_count(laliga)['GoalDifferenceCount']).idxmin()]]
```

### 5. Create a new column with name “Winning Percent” and append it to the data set
#### Percentage of Winning = (GamesWon / GamesPlayed)*100
#### If there are any numerical error, replace it with 0%
### Print the top 5 teams which has the highest Winning percentage

```python
#using assign function assign a new column 'WinningPercent' to laliga
#Covert the columns 'GamesWon' and 'GamesPlayed' to numeric using pd.numeric() and apply() functions
#Calculate Winning Percent using given formula before assigning it.
laliga = laliga.assign(WinningPercent = 
                       (laliga['GamesWon'].apply(pd.to_numeric)/laliga['GamesPlayed'].apply(pd.to_numeric))*100)
```

```python
#Check that new column is added
laliga.columns
```

```python
#Check if there are any null values in the newly added column
laliga['WinningPercent'].isna().any()
```

```python
#Using fillna(0) function, fill all null values with 0
laliga['WinningPercent'].fillna(0, inplace=True)
```

```python
#ReCheck for null values in the newly added column
laliga['WinningPercent'].isna().any()
```

```python
#Display all the required columns along with 'WinningPercent'
laliga[['Team', 'GamesPlayed', 'GamesWon', 'WinningPercent']]
```

```python
#Print the top 5 teams using sort_values() and head() functions
print('Top 5 teams having the highest Winning Percentage\n\n',
laliga[['Team', 'GamesPlayed', 'GamesWon', 'WinningPercent']].sort_values(by='WinningPercent', ascending=False).head())
```

### 6. Group teams based on their “Best position” and print the sum of their points for all positions
#### Eg: Best Position         Points
####                            1                              25000
####                            2                              7000    


```python
#Create a temp_df DataFrame with the columns 'Team', 'BestPosition', 'Points'
#Convert Points to int using astype() function and check the datatypes
temp_df = laliga[['Team','BestPosition','Points']]
temp_df = temp_df.astype({'Points':int})
temp_df.dtypes
```

```python
#Group team based on Best Position using groupby() function
#Sum the points for each Position and print all positions along with sum of points using sum() function
#convert the returned series to DataFrame using to_frame function()

print('Teams grouped based on their "Best position" and "Sum of Points for all the positions"\n\n',
      temp_df.groupby(['BestPosition'])['Points'].sum().to_frame())
```
